For a random graph the degree distribution follows a binomial distribution, but for large $N$, where $N$ is the number of entities in the network, the degree distribution follows a Poission distribution:

\begin{eqnarray}
P(k)=\exp(-\langle k\rangle)\frac{\langle k\rangle^k}{k!}.
\end{eqnarray}

$\langle k\rangle$ is the average/expected degree. In the input file, you will notice the maximum time is 100 minutes, and the rate of following for all entities is 0.1. This produces $\langle k\rangle=10.0$. Below are the plots of the data files which are also found in the directory /Random_Graph.

```python
fcumul = open('cumulative_distribution.dat', 'r')
findeg = open('in-degree_distribution.dat', 'r')
foutdeg = open('out-degree_distribution.dat', 'r')
```

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in fcumul:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("Cumulative Distribution for a Random Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("LogLog Cumulative Distribution for a Random Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

Notice that the cumulative degree distribution has a mean $k$ of 20. This is because the cumulative degree distribution takes into account of all links, both in-degree and out-degree. Plotting the in-degree and out-degree distributions we'll find the mean $k$ is 10.

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in findeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("In-degree Distribution for a Random Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("LogLog In-degree Distribution for a Random Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in foutdeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("Out-degree Distribution for a Random Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("LogLog Out-degree Distribution for a Random Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

The in-degree and out-degree distributions are identical, this is always true for a random graph when $N\rightarrow\infty$. Notice also the fall-off for all three log-log plots; this is a true indication of a random graph.
