In a scale-free network, the connections between agents come from a mechanism entitled 'preferential-attachment'. Barabasi and Albert came up with this model to describe the dynamic topology of the internet. In their model, when a new connection is made an agent will firstly look to the probability of following other existant agents based on their degree. The probability to follow an agent with degree $k_i$ is

\begin{eqnarray}
\Gamma(k_i)=\frac{k_i}{\sum_{j=1}^{N-1}k_j}.
\end{eqnarray}

The above equation normalizes the degree with the degree of every other agent ($N-1$) and this is where the probability comes from. The degree distribution has been proven analiytically by Barabasi and Albert using a continuum approach. They found that the degree distribution is approximately

\begin{eqnarray}
P(k)\approx k^{-\alpha},
\end{eqnarray}

where $\alpha =3$ for $N\rightarrow\infty$. Below are degree distributions produced by the software for 100,000 agents. The first set of plots are using an algoritm where every agent will only follow another agent once, upon joining the network. The second set of plots uses another method, discussed in the method2 directory.

```python
fcumul = open('method1/cumulative_distribution.dat', 'r')
findeg = open('method1/in-degree_distribution.dat', 'r')
foutdeg = open('method1/out-degree_distribution.dat', 'r')
```

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in fcumul:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("Cumulative Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log Cumulative Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

You can clearly see the linear relationship in the log-log plot, this gives us an easy way to find the scaling exponent of a scale-free network; we simply must find the slope of the line!

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in findeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("In-degree Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log In-degree Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

The in-degree distribution for the scale-free network looks very similar to the cumulative degree distribution above. The out-degree distribution below serves as a unit test for showing that the agents follow another agent only once.

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in foutdeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("Out-degree Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log Out-degree Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

This point ends the discussion of the degree distributions given from a scale free method with the 'use_barabasi' variable set to true. The next method of creating a scale-free network uses the same probability function to determine the protocol for following, but agents can now follow more than once. To avoid agents having an out-degree that varies greatly from the average degree, we will keep the number of agents static. This way we can have an expectation value for out-degree and this will remove the fact that agents created earlier in the simulation will have much more opportunity to follow. From the input file in the directory /method2, the expected out-degree for the simulation is 5.

```python
Image(filename='method2/scale_free_method2_viz.png')
```

The above visualization looks very similar to the first visualization with 'use_barabasi = true'. The small difference in the visualization is that there are many nodes that were not selected to follow, and therefore they have a degree of 0. The above visualization was also made using Gephi with 100 agents. The average degree for this visualization is 1.

```python
fcumul = open('method2/cumulative_distribution.dat', 'r')
findeg = open('method2/in-degree_distribution.dat', 'r')
foutdeg = open('method2/out-degree_distribution.dat', 'r')
```

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in fcumul:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("Cumulative Degree Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log Cumalitive Degree Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

You can see that for larger $k$, the log-log plot follows a (somewhat) linear relationship. These tails are called 'heavy-tails' and this is what was found in the tail of the degree distrubtion for the world wide web. This distribution does looks suspiciously like a random graph the way it falls off, below is a discussion of the reasoning behind the randomness of the distribution.

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in findeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("In-degree Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log In-degree Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

For the in-degree, we can see there is a likely probability you will have a low number of followers and very low probability you will have many followers. The linear tail found for scale free networks comes from this physical description.

```python
k = []
P_k = []
log_k = []
log_P_k = []
for line in foutdeg:
    k.append(float(line.split()[0]))
    P_k.append(float(line.split()[1]))
    log_k.append(float(line.split()[2]))
    log_P_k.append(float(line.split()[3]))
```

```python
fig,axes = subplots(1,2,figsize=(20,5))

axes[0].set_title("In-degree Distribution for a Scale-Free Graph")
axes[0].set_xlabel("k")
axes[0].set_ylabel("P(k)")
axes[0].plot(k,P_k,c='b')

axes[1].set_title("Log-Log In-degree Distribution for a Scale-Free Graph")
axes[1].set_xlabel("log[k]")
axes[1].set_ylabel("log[P(k)]")
axes[1].plot(log_k,log_P_k,c='b')
show()
```

The out degree distribution follows exactly as expected, the average number of follows an agent has is 5 (check complete). The distribution follows a Poission distribution, and this gives rise to the suspicious (fall off) tail of the cumulative distriubtion. The second method gives rise to many users not being followed. This is due to the network being static with a large number of agents. Ininitially all of the agents have 0 followers, and only a (random) select few obtain followers after some time. This creates a more random process and this can be seen in the above degree distributions. Even though following an agent has a low probability, the number of agents with 0 followers is very large, which turns following an agent with 0 followers more likely than you think.
